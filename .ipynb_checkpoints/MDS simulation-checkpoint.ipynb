## A note on the default parameters in sklearn 

It is often suboptimal to use the default sklearn parameters. There are many reasons why that's the case. For example, at least for now, when you use `LogisticRegression()` imported from `sklearn.linear_model`, by default (at least in June 2021), you are using ridge regression instead of the vanilla logistic regression, which might not be what you want. So in general, I would read the sklearn documentation page pretty carefully if you are working on a serious project. 

Here, I want to discuss some parameters that are often overlooked, among all the default parameters -- the error threshold or the stopping criterion used for deciding when to stop the optimization process during model fitting. Most sklearn methods have parameters that define the stopping criterion, so you probably used them in the past. 

In our class, we learned various properties for various machine learning algorithms, but they might not hold when you use sklearn in practice. This is partly because sklearn often approximates the solution using iterative optimization methods instead of solving for analytical solutions. 

Take multidimensional scaling (MDS), for example, which is covered this week. Though it operates on the distance matrix of the data and the distances don't change when you re-order the data, but the MDS solution can change when you re-order the data. 

Playing with the error threshold is useful here. For  [the MDS in sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.MDS.html) (the list of input arguments are listed below), 
`eps`, `n_init`, `max_iter` are parameters that affect the "error threshold" of the optimization procedure. So to get stable MDS fitting, you might want to play with these parameters. 

```python
sklearn.manifold.MDS(n_components=2, *, metric=True, n_init=4, max_iter=300, verbose=0, eps=0.001, n_jobs=None, random_state=None, dissimilarity='euclidean')
```

In the following toy simulation, you will see that the default parameters of the MDS are not strict enough, and using a more stringent criterion helped to stabilize the results. In particular, I sampled some data points from N-D space and tried to embed these points into a 2-D space. With the default parameters (e.g. `eps=0.001`), re-ordering the data changes MDS solution. When I used a more stringent criterion, the solution is more order invariant. 

This point is not specific to MDS, of course. In most cases, the default parameters in sklearn are unlikely to be optimal for your data. Similarly, playing with hyperparameter tuning is often useful. 

This is a toy simulation made for PSY/NEU/ELE 480 @ taught at Princeton Spring 2021. Comment/suggestion appreciated: qlu@princeton.edu

````python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.manifold import MDS
from sklearn.decomposition import PCA
from sklearn.metrics import euclidean_distances

sns.set(style='white', context='talk')

"""make some data"""
# params
seed = 0
n_pts = 5
n_dim = 20

np.random.seed(seed)
# make data 
X = np.random.uniform(size=(n_pts, n_dim))
# permute data 
perm_ids = np.random.permutation(n_pts)
unperm_ids = [perm_ids[i] for i in range(len(perm_ids))]
X_perm = X[perm_ids, :]

# get some colors 
cur_pals = sns.color_palette(n_colors=n_pts)
cur_pals_perm = [cur_pals[perm_ids[i]] for i in range(n_pts)]
````

````python
"""fit MDS"""
# choose some params for MDS fitting
n_inits = [4, 20]
max_iters = [300, 1000]
eps = [1e-3, 1e-10]
criterion_list = ['default criterion', 'strict criterion']
print('MDS - Embed %d points from %d-D space to 2D ' % (n_pts, n_dim))

# loop over different stopping criteria 
alpha = 1
length = 3
for i, criterion in enumerate(criterion_list):
    # fit MDS on the raw data
    mds = MDS(
        n_components=2, random_state=seed, 
        eps=eps[i], n_init=n_inits[i], max_iter=max_iters[i],

    )
    Y = mds.fit_transform(X)
    # fit MDS on the permuted data
    mds_perm = MDS(
        n_components=2, random_state=seed, 
        eps=eps[i], n_init=n_inits[i], max_iter=max_iters[i],
    )
    Y_perm = mds_perm.fit_transform(X_perm)

    """plot the solution"""
    
    f, axes = plt.subplots(1, 2, figsize=(9, 4))
    axes[0].plot(Y[:,0], Y[:,1], color='grey', alpha=alpha, linestyle = '--')
    axes[0].scatter(Y[:,0], Y[:,1], c = cur_pals)
    axes[0].set_title(f'raw, {criterion}')

    Y_unperm = Y_perm[unperm_ids]
    axes[1].plot(Y_unperm[:,0],Y_unperm[:,1], color='grey', alpha=alpha, linestyle = '--')
    axes[1].scatter(Y_unperm[:,0], Y_unperm[:,1], c = cur_pals)
    axes[1].set_title(f'permuted, {criterion}')    
    for i in range(len(axes)):
        axes[i].axis('equal')
        axes[i].axis([-length, length, -length, length])    
    f.tight_layout()
    sns.despine()
````

In the lab notebook, there is a toy simulation that embeds a line into N-D space. The result shows that as N gets bigger, the stress increases. Why is that? Well, before you think too hard, note that if we use a strict stopping criterion, that pattern no longer holds... Can you explain why? 

````python
# sample data 
n_pts = 30
coords = np.linspace(1, n_pts, n_pts)
coords = np.vstack((coords, coords, coords)).T

distance_matrix = euclidean_distances(coords)
n_components_list = [1, 2, 4, 8, 16]

# plot the stress over the dimension of MDS
Y = {c: [] for c in criterion_list}
f, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
for i, criterion in enumerate(criterion_list):
    mds_list = []
    stress = []
    for k in n_components_list:
        mds = MDS(
            n_components=k, dissimilarity="precomputed", random_state=seed,
            eps=eps[i], n_init=n_inits[i], max_iter=max_iters[i],
        )
        Y[criterion].append(mds.fit_transform(distance_matrix))
        stress.append(mds.stress_)

    axes[i].plot(n_components_list, stress, marker = 'o')
    axes[i].set_title(f'{criterion}')
    axes[i].set_ylabel('Stress')
    axes[i].set_xlabel('MDS Dimension')
    sns.despine()
f.suptitle(f'Embed a 3d line to N-D space')
f.tight_layout()    

# plot the 2D solution 
f, axes = plt.subplots(1, 2, figsize=(12, 5))
k_id = 1
for i, criterion in enumerate(criterion_list):
    axes[i].scatter(Y[criterion][k_id][:,0], Y[criterion][k_id][:,1])
    axes[i].set_title(criterion)
    sns.despine()
f.suptitle(f'2D solution')    
f.tight_layout()
````

Just another bonus point here since you have seen both PCA and MDS. I'm embedding the some data with PCA and MDS, and plotting them side by side. What do you see? Do some research to find out why :) 

````python
# sample data
n_pts = 6
X = np.random.uniform(size=(n_pts, n_dim))
cur_pals = sns.color_palette(n_colors=n_pts)

# embed the data to 2d space
pca = PCA(n_components=2,random_state=seed)
X_pc = pca.fit_transform(X)

mds = MDS(n_components=2, random_state=seed)
X_mds = pca.fit_transform(X)

# plot the embeddings
dim_reduced_data = [X_pc, X_mds]
names = ['PCA', 'MDS']
f,axes = plt.subplots(1,2, figsize=(9,4))
for i, X_ in enumerate(dim_reduced_data): 
    axes[i].plot(X_[:,0], X_[:,1], color='grey', alpha=alpha, linestyle = '--')
    axes[i].scatter(X_[:,0], X_[:,1], color = cur_pals)    
    axes[i].set_title(names[i])
f.tight_layout()
sns.despine()
````
